```python
#seleziono solo i file .csv.zst
files_csv=[]
for f in files:
    if f.endswith('.csv.zst'):
        files_csv.append(f)
files_csv
```

```python
#unisco i file in un unico dataframe
df = pd.read_csv(folder+files_csv[0],encoding='latin-1')
df["Country"]=files_csv[0][:2]

for f in range(1,len(files_csv)):
    ds = pd.read_csv(folder+files_csv[f],encoding="latin-1")
    ds["Country"]=files_csv[f][:2]
    df=pd.concat([df,ds], ignore_index=True)

display(df.head(10))
```

4. Save all rows with disabled comments and disabled ratings, or that have video_error_or_removed in a new dataframe called excluded, and remove those rows from the original dataframe.

```python
excluded = df[((df['ratings_disabled'] == True) & (df['comments_disabled'] == True)) 
   | (df['video_error_or_removed'] == True)]
excluded
# rimuovo righe dal dataframe originale
df = df.drop(excluded.index)
df.head(10)
```

5. Add a like_ratio column storing the ratio between the number of likes and of dislikes


```python
df['like_ratio']=round(df['likes']/df['dislikes']*100,2)
df[['video_id', 'views', 'likes', 'dislikes', 'like_ratio']].head(10)
```

6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
df['publish_time_10min'] = pd.to_datetime(df['publish_time']).dt.floor('10min')
df['interval'] = df['publish_time_10min'].dt.strftime('%H:%M') + '-' + \
                  (df['publish_time_10min'] + pd.Timedelta(minutes=10)).dt.strftime('%H:%M')
df[['video_id','publish_time', 'publish_time_10min', 'interval']].head(10)
```

7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python
dfg2 = df.groupby('interval').agg({
    'video_id': 'count',      # numero di video
    'likes': 'mean',          # media like
    'dislikes': 'mean'        # media dislike
}).reset_index()
#arrotondo le medie
dfg2['likes']= round(dfg2['likes'],2)
dfg2['dislikes']= round(dfg2['dislikes'],2)
dfg2.rename(columns={'video_id': 'videos num', 'likes': 'likes avg', 'dislikes': 'dislikes avg'}, inplace=True)

dfg2
```

8. For each tag, determine the number of videos
Notice that tags contains a string with several tags.

```python
# analisi tag
dft=df
dft['tags'] = dft['tags'].str.split('|')
#normalizzo i tag
tag_counts = (
    dft.explode('tags')
      .assign(tags=lambda x: x['tags']
              .str.replace('"', '', regex=False)  # remove quotes
              .str.strip())                       # remove spaces
      .groupby('tags')
      .size()
      .reset_index(name='video_count')
      .sort_values('video_count', ascending=False)
      .reset_index(drop=True)
)
tag_counts = tag_counts[tag_counts['tags'] != '[none]']
display(tag_counts.head(10))


```

9. Find the tags with the largest number of videos

```python
top_tag = tag_counts.head(1)
display(top_tag)
```

10. For each (tag, country) pair, compute average ratio likes/dislikes

```python
#copia del dataframe originale
df_tags = df.copy()

df_tags["tags_clean"] = df_tags["tags"].astype(str).str.strip().str.replace('"', '', regex=False)

df_tags["tag"] = df_tags["tags_clean"].str.split("|")

df_tags = df_tags.explode("tag")

#rimuovo spazi residui e tag vuoti
df_tags["tag"] = df_tags["tag"].str.strip()
df_tags = df_tags[df_tags["tag"].notna() & (df_tags["tag"] != "")]

#calcolo la media like_ratio per (tag, Country)
result = (
    df_tags
    .groupby(["tag", "Country"])["like_ratio"]
    .mean()
    .reset_index(name="avg_like_ratio")
    .sort_values(by="avg_like_ratio", ascending=False)
)

display(result)
```

14. Read all json files with the video categories

```python
path = folder

#prendo tutti i file *_category_id.json
files = glob.glob(os.path.join(path, "*_category_id.json"))

dfs = []
#apro ogni file json 
for file in files:
    
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)

    #normalizzo la lista "items"
    dfc = pd.json_normalize(data["items"])

    #aggiungo il nome del file
    dfc["source_file"] = os.path.basename(file)

    dfs.append(dfc)

#unisco tutti i dataframe dei file json 
final_df = pd.concat(dfs, ignore_index=True)

#estraggo il country code dal nome file 
final_df["country"] = final_df["source_file"].str[:2]

final_df = final_df.rename(columns={
    "snippet.title": "category_name",
    "snippet.assignable": "assignable"
})

display(final_df)
```

15. For each country, determine how many videos have a category that is not assignable.

```python
#merge tra video e categorie usando i nomi originali
df['category_id'] = df['category_id'].astype(int)
final_df['id'] = final_df['id'].astype(int)
merged_df = df.merge(
    final_df[['id', 'assignable', 'country']],
    left_on=['category_id', 'Country'],
    right_on=['id', 'country'],
    how='left'
)

#filtro solo le categorie non assignable
non_assignable_df = merged_df[merged_df['assignable'] == False]

#conto il numero di video non assignable per Paese
result = non_assignable_df.groupby('Country')['video_id'].count().reset_index(name='num_non_assignable_videos')

display(result)


```
